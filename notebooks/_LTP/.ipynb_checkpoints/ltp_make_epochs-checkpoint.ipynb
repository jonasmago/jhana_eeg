# Import 

```python
# vertical tetanus: 0,1,8,9
# Vertical = White = 1
# Horizontal = Black = -1

# I leave the numbers as are for horizontal tetanus. I invert the numbers for vertical tetanus. 
# In this case, it's always -1 that represents the tetanized version. 

# --> +1 = NORMAL
# --> -1 = TETANUS 

def raw_to_events(raw, timepoint, condition, sub):
    if sub in ('sub2', 'sub3', 'sub4', 'sub5', 'sub6', 'sub7'):
        # tetanus is HORIZONTAL
        inversion = 1
    if sub in ('sub0', 'sub1', 'sub8', 'sub9'):
        # tetanus is VERTICAL
        inversion = -1

    '''
    if timepoint == 'pre' and condition == 'baseline':
        event_id = {'-1':-1*inversion, '1':1*inversion,}
    elif timepoint == 'pre' and condition == 'early':
        event_id = {'-1':-2*inversion, '1':2*inversion,}
    elif timepoint == 'pre' and condition == 'late':
        event_id = {'-1':-3*inversion, '1':3*inversion,}
    elif timepoint == 'post' and condition == 'baseline':
        event_id = {'-1':-4*inversion, '1':4*inversion,}
    elif timepoint == 'post' and condition == 'early':
        event_id = {'-1':-5*inversion, '1':5*inversion,}
    elif timepoint == 'post' and condition == 'late':
        event_id = {'-1':-6*inversion, '1':6*inversion,}
    '''
    
    event_id = {'-1':-1*inversion, '1':1*inversion,}
    
    events = mne.events_from_annotations(raw, event_id=event_id)[0]
    
    if len(events)==0:
        events = np.zeros((1,3))
        events = events.astype(int)
    return events 
```

```python
paths = glob.glob(f"../../notebooks/hand_cleaning/ltp/sub?-LTP_*-*_ltp-raw_phot-events.fif")
paths.sort()

for path in paths:
    filename_components = re.split(r'[-_]', path)
    condition = filename_components[-5]
    timepoint = filename_components[-6]
    sub = filename_components[-8][-4:]

    raw = mne.io.read_raw_fif(path, preload=True)
    montage = mne.channels.make_standard_montage('standard_1020')
    raw.set_montage(montage)    
    
    events = raw_to_events(raw, timepoint, condition, sub)   
    epochs = mne.Epochs(raw, events, tmin=-0.2, tmax=0.5, baseline=(-0.2, 0), preload=True)

    if len(epochs) < 10:
        print (f'not enough epochs for {path}')
        continue 

    #reject_criteria = dict(eeg=100e-6)
    #epochs.drop_bad(reject=reject_criteria)
    
    basename = os.path.basename(path)
    export_path = f"epochs/{sub}_{timepoint}_{condition}_epo.fif"
    epochs.save(export_path, overwrite=True)
```

## ICA

```python
sub = 4
time = 'post' #'pre'

paths = glob.glob(f'epochs/sub{sub}_{time}_*_epo.fif')
paths.sort()

epochs_list = [mne.read_epochs(path, preload=True) for path in paths]

for i, epochs in enumerate(epochs_list):
    epochs.interpolate_bads(reset_bads=True).pick_types(eeg=True).filter(l_freq=None, h_freq=40)
    epochs_list[i] = epochs  # Explicitly store the updated object back into the list

combined_epochs = mne.concatenate_epochs(epochs_list)
#combined_epochs.filter(l_freq=1., h_freq=40.)

ica = ICA(n_components=20, random_state=97, max_iter=800)
ica.fit(combined_epochs)

ica.plot_sources(combined_epochs)  # Plots the time series of all ICA components
ica.plot_components(picks=None)  # 'picks=None' shows all components

print('')
```

```python
plt.close('all')
for i, epochs in enumerate(epochs_list):
    ica.exclude = [0]
    cleaned_epochs = ica.apply(epochs.copy())
    input_path = paths[i]
    output_path = input_path.replace('epochs', 'epochs_ica')
    cleaned_epochs.save(output_path, overwrite=True)
```
