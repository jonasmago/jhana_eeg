```python
%load_ext autoreload
%autoreload 2
%matplotlib qt

import numpy as np
import autoreject
import csv
import matplotlib.pyplot as plt
import pandas as pd
import mne
import os
import sys
from atpbar import atpbar
import glob
from autoreject import AutoReject
from mne.preprocessing import ICA, create_eog_epochs, create_ecg_epochs
```

# AR cleaning

```python
import_folder = "../../data/mne_raw_epochs/"
export_folder = "../../data/mne_raw_epochs_ar/"
    
conditions = ["jhana_mmn",
              "mindfulness_mmn",
              "control_mmn",
              "rmmn",
              "ltp",
             "rs_control_counting",
             "rs_control_mindfulness",
             "rs_eyes_closed_post",
             "rs_eyes_closed_pre",
             "rs_eyes_open_post",
             "rs_eyes_open_pre",
             "rs_jhana",
             "rs_mindfulness"]

warnings = []

for condition in conditions: 
    if not os.path.exists(os.path.join(export_folder, condition)):
        os.makedirs(os.path.join(export_folder, condition))
    search_string = os.path.join(import_folder, condition, "*")
    paths = glob.glob(search_string)
    paths.sort()
    for path in paths: 
        try: 
            epochs = mne.read_epochs(path)
            ar = AutoReject(picks="eeg")
            epochs_ar_clean, reject_log = ar.fit_transform(epochs, return_log=True)
            epochs_ar_clean.set_eeg_reference(ref_channels='average')
            export_path = os.path.join(export_folder, condition, os.path.basename(path))
            epochs_ar_clean.save(export_path, overwrite=True)
            
            reject_log.plot('horizontal')
            plt.savefig(export_path[:-4]+".png")
            plt.close()
            
        except: 
            warnings.append(path)
            print (f"warning! {path}")
```

# ICA and stuff

```python
# clean EPOCHS

def clean_epochs_full (epochs):
    ica = mne.preprocessing.ICA(n_components=32, random_state=97, max_iter="auto")
    ica.fit(epochs)
    muscle_idx_auto, scores = ica.find_bads_muscle(epochs)
    print (f"excluded ICA components: {muscle_idx_auto}")
    epochs_ica = ica.apply(epochs.copy(), exclude=muscle_idx_auto)
    
    ar = autoreject.AutoReject(n_interpolate=[1, 2, 3, 4], random_state=11,n_jobs=1, verbose=False)
    ar.fit(epochs_ica)
    epochs_ar, reject_log = ar.transform(epochs_ica, return_log=True)
    epochs_ar.filter(None, 40)
    print (f"retention rate: {len(epochs_ar)}/{len(epochs)}")
    return (epochs, epochs_ica, epochs_ar, reject_log)
```
