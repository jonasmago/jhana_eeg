```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = df_combined.columns[8:]
corr_matrix = pd.DataFrame(index=numeric_columns, columns=numeric_columns)
pval_matrix = pd.DataFrame(index=numeric_columns, columns=numeric_columns)

# Apply mixed-effects modeling for each pair
for col1 in numeric_columns:
    for col2 in numeric_columns:
        if col1 == col2:
            corr_matrix.loc[col1, col2] = 1.0
            pval_matrix.loc[col1, col2] = 0.0
        else:
            # Drop rows with missing values in either column
            df_clean = df_combined.dropna(subset=[col1, col2])
            if len(df_clean) > 1:
                # Define and fit the mixed-effects model
                model = smf.mixedlm(f"{col1} ~ {col2}", df_clean, groups=df_clean["sub"], re_formula="~1")
                result = model.fit()
                
                # Access slope and p-values using iloc for positional indexing
                corr_matrix.loc[col1, col2] = result.params.iloc[1]  # Use iloc for positional access
                pval_matrix.loc[col1, col2] = result.pvalues.iloc[1]  # Use iloc for positional access
            else:
                corr_matrix.loc[col1, col2] = np.nan
                pval_matrix.loc[col1, col2] = np.nan
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

# Select the 33 numeric columns from df_combined
numeric_columns = df_combined.columns[8:]

# Initialize matrices for storing correlation coefficients and p-values
corr_matrix = pd.DataFrame(index=numeric_columns, columns=numeric_columns)
pval_matrix = pd.DataFrame(index=numeric_columns, columns=numeric_columns)

# Compute correlations and p-values, excluding missing values for each pair
for col1 in numeric_columns:
    for col2 in numeric_columns:
        if col1 == col2:
            corr_matrix.loc[col1, col2] = 1.0
            pval_matrix.loc[col1, col2] = 0.0
        else:
            # Drop rows with NaN values in either column
            valid_data = df_combined[[col1, col2]].dropna()
            if len(valid_data) > 1:  # Check that there is more than 1 valid pair
                corr, pval = pearsonr(valid_data[col1], valid_data[col2])
                corr_matrix.loc[col1, col2] = corr
                pval_matrix.loc[col1, col2] = pval
            else:
                corr_matrix.loc[col1, col2] = np.nan
                pval_matrix.loc[col1, col2] = np.nan

# Convert the matrices to numeric (since they are initialized with NaNs)
corr_matrix = corr_matrix.astype(float)
pval_matrix = pval_matrix.astype(float)

# Create a mask for the upper triangle of the matrix (to avoid duplication in the plot)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot the correlation matrix heatmap
plt.figure(figsize=(30, 30))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, cbar_kws={'label': 'Correlation Coefficient'})
plt.title("Correlation Coefficient Matrix")

# Show the plot
plt.tight_layout()
plt.show()

# Plot the significance (p-value) matrix heatmap with two decimal places for p-values
plt.figure(figsize=(30, 30))
sns.heatmap(pval_matrix, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={'label': 'p-value'})
plt.title("P-value Matrix (Significance of Correlations)")

# Show the plot
plt.tight_layout()
plt.show()
```
