```python
import mne
import numpy as np
import matplotlib.pyplot as plt

# Define frequency bands
frequency_bands = {
    'delta': (0.5, 4),
    'theta': (4, 8),
    'alpha': (8, 12),
    'beta': (12, 30),
    'gamma': (30, 45)
}

# Compute PSD using Welch's method for all epochs
psd = epochs.compute_psd(fmin=0.5, fmax=45, n_fft=256, method='welch')
psds, freqs = psd.get_data(return_freqs=True)

# Average the PSD across all epochs
psds_mean = np.mean(psds, axis=0)  # Shape will be (n_channels, n_freqs)

# Pick the good channels (exclude the bad channels)
good_channels = mne.pick_info(epochs.info, sel=mne.pick_types(epochs.info, eeg=True, exclude='bads'))

# Prepare the figure for the topomaps
fig, axes = plt.subplots(1, len(frequency_bands), figsize=(15, 4))  # Create a subplot for each band

# Loop over each frequency band and plot a topomap
for i, (band, (fmin, fmax)) in enumerate(frequency_bands.items()):
    # Find the frequency indices corresponding to the band
    freq_mask = (freqs >= fmin) & (freqs < fmax)

    # Compute the average power in the band
    band_power = np.mean(psds_mean[:, freq_mask], axis=1)

    # Create a topomap for this band using only good channels
    mne.viz.plot_topomap(band_power, good_channels, axes=axes[i], show=False, cmap='viridis')
    axes[i].set_title(f'{band} band ({fmin}-{fmax} Hz)')

# Show the final image with all topomaps
plt.tight_layout()
plt.show()
```

```python
import mne
import numpy as np

# Define the electrode names and corresponding values manually
channels = ['Fp1', 'Fp2', 'Fz', 'Cz', 'Pz', 'Oz', 'T3', 'T4', 'O1', 'O2']
values = [1, 0.15, 0.5, 0.2, 0.35, 0.1, 0.05, 0.45, 0.25, 0.3]  # Example values

# Create an MNE Info object with the standard 10-20 montage
montage = mne.channels.make_standard_montage('standard_1020')
info = mne.create_info(ch_names=channels, sfreq=256, ch_types='eeg')
info.set_montage(montage)

# Create a fake Evoked object using the values (as a single time point)
evoked = mne.EvokedArray(np.expand_dims(values, axis=1), info)

# Plot the topomap
mne.viz.plot_topomap(evoked.data[:, 0], evoked.info, show=True, cmap='viridis')
```

```python
#paths = glob.glob('epochs/*')
paths = glob.glob('epochs_ar/*')
paths.sort()
paths = paths

frequency_bands = {
    'delta': (0.5, 4),
    'theta': (4, 8),
    'alpha': (8, 12),
    'beta': (12, 30),
    'gamma': (30, 45)
}


for path in paths: 

    # naming all the components
    #components = os.path.basename(path).split('_')
    components = re.split('[-_]', os.path.basename(path))
    sub = components[0][3:]
    day = components[1]
    condition = components[2]
    if day == "control":
        sub = 'control_'+sub
    
    #analysis
    epochs = mne.read_epochs(path)

    
    if len(epochs) < 10: 
        lempel_ziv_norm = np.nan
        permutation_entropy = np.nan
        spectral_entropy = np.nan
        sample_entropy = np.nan
        hjorth_mobility = np.nan
        hjorth_complexity = np.nan
        average_corr_all = np.nan
        band_powers_mean = {band: np.nan for band in frequency_bands}
    
    
    else:   
        data = epochs.get_data(copy=False)
        
        lempel_ziv_norm = []
        permutation_entropy = []
        spectral_entropy = []
        sample_entropy = []
        hjorth_mobility = []
        hjorth_complexity = []
        average_corr_all = []
    
        band_powers = {band: [] for band in frequency_bands}
    
    
        n_channels = 32 - len(epochs.info['bads'])   
        print (f'path: {path}')
        print(f'n_channels: {n_channels}')
        print (f'number epochs: {len(data)}')
        for epoch in range(len(data)):
            for channel in range(n_channels):
                time_series = data[epoch,channel,:]
                threshold = np.median(time_series)
                binary_sequence = (time_series > threshold).astype(int)
                binary_string = ''.join(binary_sequence.astype(str))
        
                
                lempel_ziv_norm.append(ant.lziv_complexity(binary_string, normalize=True))
                permutation_entropy.append(ant.perm_entropy(time_series, normalize=True))
                spectral_entropy.append(ant.spectral_entropy(time_series, sf=100, method='welch', normalize=True))
                sample_entropy.append(ant.sample_entropy(time_series))
                mobility, complexity = ant.hjorth_params(time_series)
                hjorth_mobility.append(mobility)
                hjorth_complexity.append(complexity)
                
    
            correlation_matrix = np.corrcoef(data[epoch,:32,:])
            upper_triangle_indices = np.triu_indices(32, k=1)
            average_corr = np.mean(correlation_matrix[upper_triangle_indices])
            average_corr_all.append(average_corr)
    
            
            psd, freqs = mne.time_frequency.psd_array_welch(time_series, sfreq=epochs.info['sfreq'], fmin=0.5, fmax=45, n_fft=256)
        
            for band, (fmin, fmax) in frequency_bands.items():
                band_power = np.mean(psd[(freqs >= fmin) & (freqs < fmax)])
                band_powers[band].append(band_power*1e12)
        
        lempel_ziv_norm = np.round(np.mean(lempel_ziv_norm), 4)
        permutation_entropy = np.round(np.mean(permutation_entropy), 4)
        spectral_entropy = np.round(np.mean(spectral_entropy), 4)
        sample_entropy = np.round(np.mean(sample_entropy), 4)
        hjorth_mobility = np.round(np.mean(hjorth_mobility), 4)
        hjorth_complexity = np.round(np.mean(hjorth_complexity), 4)
        average_corr_all = np.round(np.mean(average_corr_all), 4)
        band_powers_mean = {band: np.round(np.mean(power), 4) for band, power in band_powers.items()}


    df_data = {
        'sub': sub, 
        'day': day, 
        'condition': condition, 
        
        'lempel_ziv_norm': lempel_ziv_norm,
        'permutation_entropy': permutation_entropy,
        'spectral_entropy': spectral_entropy,
        'sample_entropy': sample_entropy,
        'hjorth_mobility': hjorth_mobility,
        'hjorth_complexity': hjorth_complexity,
        'average_corr': average_corr_all,
                }
    df_data.update(band_powers_mean)

    df_row = pd.DataFrame([df_data])
    
    if path == paths[0]:
        df_results = df_row
    else:
        df_results = pd.concat([df_results, df_row], ignore_index=True)
    df_results.to_csv('eeg_metrics_hand.csv', index=False)
    
```
